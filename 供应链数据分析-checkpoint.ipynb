1.产品分析

```python
product_counts = df["Product type"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(product_counts,
        labels=product_counts.index,
        autopct='%1.1f%%')

plt.title("Product Type Distribution")
plt.show()
```

根据以上饼图可以明显看出在产品类型这部分销量占比护肤品最多，之后是护发品，最后才是化妆品

```python
#1.按 SKU 汇总收入
revenue_by_sku = (
    df.groupby("SKU", as_index=False)["Revenue generated"]
      .sum()
      .sort_values("Revenue generated", ascending=False)
)
print(revenue_by_sku.head(10))

fig_revenue = px.bar(
    revenue_by_sku,
    x="Revenue generated",
    y="SKU"  , # 横着画
    title="Total Revenue by SKU"
)


fig_revenue.update_layout(
    xaxis_title="Total Revenue",
    yaxis_title="SKU"
)
fig_revenue.show()
```

```python
#2.按 SKU 汇总库存
stock_by_sku = (
    df.groupby("SKU", as_index=False)["Stock levels"]
      .sum()
      .sort_values("Stock levels", ascending=False)
)
print(stock_by_sku.head(10))
```

```python
#3.按 SKU 汇总订单数量
orders_by_sku = (
    df.groupby("SKU", as_index=False)["Order quantities"]
      .sum()
      .sort_values("Order quantities", ascending=False)
)
print(orders_by_sku.head(10))
```

```python
#4.按 SKU计算利润
df["Profit"] = df["Revenue generated"] - df["Costs"]

profit_by_sku = (
    df.groupby("SKU", as_index=False)["Profit"]
      .sum()
      .sort_values("Profit", ascending=False)
)
print(profit_by_sku.head(10))
```

结论：1.收入排名靠前的SKU与利润排名高度重合，说明当前高收入产品同时也是主要利润来源，产品结构较为健康。
      2.部分SKU库存排名靠前但未进入收入前列，说明存在库存积压风险，库存配置与销售表现不完全匹配
      3.个别SKU订单量较高但未进入收入前列，可能存在单价较低或盈利能力较弱的问题。
      4.多个SKU如SKU32、SKU0、SKU88、SKU98、SKU61、SKU18、SKU85，在收入、订单和利润中均表现突出，是当前业务的核心产品，应重点保障库存与供应，优化供应链配置

2.客户结构

```python
demo_counts = df["Customer demographics"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(demo_counts,
        labels=demo_counts.index,
        autopct='%1.1f%%')

plt.title("Customer Demographics Distribution")
plt.show()
```

结论：公司的客户构成有男性、女性、非二元性别和未知性别群体，因此公司在制定营销策略时，
      需要注意营销策略应该更前卫包容涵盖所有客户群体

```python
#1.公司在不同客户群体的平均收入
revenue_by_demo = (
    df.groupby("Customer demographics")["Revenue generated"]
    .mean()
    .sort_values(ascending=False)
)
print(revenue_by_demo)

revenue_by_demo.plot(kind="bar")
plt.title("Average Revenue by Customer Demographics")
plt.ylabel("Revenue generated")
plt.xlabel("Customer demographics")
plt.xticks(rotation=45)
plt.show()
```

结论：公司在女性客户群体所获得的收入明显高于其他群体，因此公司营销活动时需要注意女性客户群体对于营销活动的反馈便于及时调整

3.定价与收入分析

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Product type", y="Price")
plt.title("Distribution of Price")
plt.xlabel("Product type")
plt.ylabel("Price")
plt.show()
```

结论：通过箱线图可以观察到，不同产品类型之间的价格分布存在明显差异。
     从价格分布来看，化妆品的价格主要集中在40–80区间，整体分布较为集中，说明其产品定价较为统一。
     而护肤品的价格范围更广（20–80），既包含低价产品，也包含中等价位产品，说明其产品线更加多样化。
     这表明不同产品类型在定价策略上存在差异，护肤品更倾向于覆盖不同价格层级的市场，而化妆品则更集中于某一价格区间。

4.运输分析

```python
#1.不同运输方式的平均运输成本
shipping_cost_by_mode = (
    df.groupby("Transportation modes")["Shipping costs"]
    .mean()
    .sort_values(ascending=False)
)
print(shipping_cost_by_mode)

shipping_cost_by_mode.plot(kind="bar")
plt.title("Average Shipping Costs by Transportation Modes")
plt.ylabel("Shipping costs")
plt.xlabel("Transportation modes")
plt.xticks(rotation=45)
plt.show()
```

结论：不同运输方式的平均运输成本有差别，运输成本由高到低是空运、铁路、公路、海运

```python
#2.不同运输方式的平均时间
shipping_time_by_mode = (
    df.groupby("Transportation modes")["Shipping times"]
    .mean()
    .sort_values(ascending=False)
)
print(shipping_time_by_mode)
```

结论：从不同运输方式平均成本上来讲，运输成本由高到低是空运、铁路、公路、海运。
      当从不同运输方式的平均运输时间上来讲，运输时间由长到短是海运、铁路、空运、公路。
      因此从运输成本和运输时间上来讲，更多的考虑公路运输可以提高消费者消费体验并降低成本

```python
#3.不同产品类型经过运输后的不良率 
shipping_rates_by_type = (
    df.groupby("Product type")["Defect rates"]
    .mean()
    .sort_values(ascending=False)
)
print(shipping_rates_by_type)
```

结论：护发产品的运输不良率最高，可以改进护发产品的运输包装进而改善护发产品的运输不良率

```python
#4.不同运输方式的不良率
shipping_rates_by_mode = (
    df.groupby("Transportation modes")["Defect rates"]
    .mean()
    .sort_values(ascending=False)
)
print(shipping_rates_by_mode)
```

结论：运输成本由高到低是空运、铁路、公路、海运，但是运输方式的不良率上由高到低是公路、海运、铁路、空运。

```python
#5不同路线的平均成本
cost_by_route = (
    df.groupby("Routes")["Costs"]
    .mean()
    .sort_values(ascending=False)
)
print(cost_by_route)

#plt.figure(figsize=(10, 5))
#cost_by_route.plot(kind="bar")
#plt.title("Average Costs by Routes")
#plt.ylabel("Costs")
#plt.xlabel("Routes")
#plt.xticks(rotation=45)
#plt.show()
```

5.供应商交货时间和运输成本分析

```python
supplier_summary = (
    df.groupby("Supplier name")
    .agg({
        "Lead time": "mean",
        "Shipping costs": "mean"
    })
    .sort_values(by="Lead time", ascending=False)
)

supplier_summary
```

结论:不同供应商的平均交货时间存在差异，从供应商的交货时间长短来看，从长到短供应商3、供应商2、供应商5、供应商4、供应商1。
     公司可以更多地采购供应商1和供应商4的货物
